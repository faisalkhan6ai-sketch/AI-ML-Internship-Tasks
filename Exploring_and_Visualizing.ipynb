```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
iris=pd.read_csv("/content/Iris.csv")
#shape and columns
print(iris.shape)
print(iris.columns.tolist())
#Show first 5 rows
print(iris.head())
#Show summary info
print(iris.info())
#show Statistics
print(iris.describe())
#create a scatter chart of Sepal Length and Sepal Width
sns.scatterplot(x="SepalLengthCm",y="SepalWidthCm",hue="Species",data=iris)
#Give title to scatter chart
plt.title("Sepal Length vs Sepal Width")
#show the chart`
plt.show()
#create a scatter chart of Petal Length  and Petal Width
sns.scatterplot(x="PetalLengthCm",y="PetalWidthCm",hue="Species",data=iris)
plt.title("Petal Length vs Petal Width")
plt.show()
#create a histogram of Sepal Length
sns.histplot(x="SepalLengthCm",hue="Species",data=iris)
plt.title("SepalLengthCm")
plt.show()
#create a histogrm of sepal width
sns.histplot(x="SepalWidthCm",hue="Species",data=iris)
plt.title("SepalWidthCm")
plt.show()
#create a histogram of petal length
sns.histplot(x="PetalLengthCm",hue="Species",data=iris)
plt.title("PetalLengthCm")
plt.show()
#create a histogram of petal width
sns.histplot(x="PetalWidthCm",hue="Species",data=iris)
plt.title("PetalWidthCm")
plt.show()
#create a box chart
sns.boxplot(x="Species",y="SepalLengthCm",data=iris)
plt.title("SepalLengthCm")
plt.show()
#create a box chart
sns.boxplot(x="Species",y="SepalWidthCm",data=iris)
plt.title("SepalWidthCm")
plt.show()
sns.boxplot(x="Species",y="PetalLengthCm",data=iris)
plt.title("PetalLengthCm")
plt.show()
sns.boxplot(x="Species",y="PetalWidthCm",data=iris)
plt.title("PetalWidthCm")
plt.show()
```
